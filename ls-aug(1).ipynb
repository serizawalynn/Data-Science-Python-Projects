## Create the training and testing dataframes.

```python
from sklearn.model_selection import train_test_split
features=list(df.columns)
features.remove("loss")
features.remove("id")
x_train,x_test,y_train,y_test=train_test_split(df[features],df["loss"],test_size=.2,random_state=1)
```

#### Check to make sure that the features we want are included. 

```python
print(features)
```

## Scaling the data.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('normalize',StandardScaler(),features)],remainder="passthrough")
x_train=ct.fit_transform(x_train)
x_test=ct.transform(x_test)
```

## Modeling
I chose the following models:
1. XGB
2. SVR where kernel=linear
3. SVR where kernel=rbf
4. Ridge Regression
5. Linear Regression

These models were chosen based on sklearn's flowchart for models. 

```python
#XGB
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
xgb_model=XGBRegressor(n_jobs=4)
xgb_model.fit(x_train,y_train)
xgb_preds=xgb_model.predict(x_test)
xgb_score=mean_squared_error(y_test,xgb_preds,squared=False)
print(xgb_score)
```

```python
#ARD Regression
from sklearn.linear_model import ARDRegression
ard_model=ARDRegression()
ard_model.fit(x_train,y_train)
ard_model_preds=ard_model.predict(x_test)
ard_model_score=mean_squared_error(y_test,ard_model_preds,squared=False)
print(ard_model_score)
```

```python
#SGD Regressor
from sklearn.linear_model import SGDRegressor
sgd_model=SGDRegressor()
sgd_model.fit(x_train,y_train)
sgd_model_preds=sgd_model.predict(x_test)
sgd_score=mean_squared_error(y_test,sgd_model_preds,squared=False)
print(sgd_score)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log_reg_model=LogisticRegression()
log_reg_model.fit(x_train,y_train)
log_reg_model_preds=log_reg_model.predict(x_test)
log_reg_score=mean_squared_error(y_test,log_reg_model_preds,squared=False)
print(log_reg_score)
```

```python
#Linear SVR
from sklearn.svm import LinearSVR
lin_svr_model=LinearSVR()
lin_svr_model.fit(x_train,y_train)
lin_svr_model_preds=lin_svr_model.predict(x_test)
lin_svr_model_score=mean_squared_error(y_test,lin_svr_model_preds,squared=False)
print(lin_svr_model_score)
```

```python
#Ridge Regression
from sklearn import linear_model
linreg_model=linear_model.Ridge()
linreg_model.fit(x_train,y_train)
linreg_model_preds=linreg_model.predict(x_test)
linreg_model_score=mean_squared_error(y_test,linreg_model_preds,squared=False)
print(linreg_model_score)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
regression_model=LinearRegression()
regression_model.fit(x_train,y_train)
regression_model_preds=regression_model.predict(x_test)
regression_model_score=mean_squared_error(regression_model_preds,y_test,squared=False)
print(regression_model_score)
```

Based on the above scores (or, what I got when I ran them), the XGB Regressor seems to be the most promising. I will now optimize it using optuna.
I went from:
* L_R: .01 to .02
* n_estimators:1000 to 3000

Then:
* L_R: .01 to .015
* n_estimators: 2000 to 3000

Next:
* L_R: .01 to .0125
* n_estimators: 2250 to 2750

Afterward:
* L_R: .01 to .011
* n_estimators: 2250 to 2500

I am adjusting the optuna.trial.suggest_uniform values in this way because I am trying to decrease the interval where I believe the most optimal parameters are. This relies on a few assumptions:
* The optimal parameters are in the initial interval
* The optimal parameters are not in the difference between the intervals
  
  For example: I am assuming that the best n_estimators is not between 1000 and 2000, which was removed from the first to second intervals.

```python
#Optimization via Optuna
import optuna
def objective(trial):
    #Parameter Suggestions
    L_R=trial.suggest_uniform('L_R',.01,.011)
    N_E=trial.suggest_int('N_E',2250,2500)
    
    
    #Machine learning logic
    model=XGBRegressor(learning_rate=L_R,n_jobs=4,n_estimators=N_E)
    model.fit(x_train,y_train)
    preds=model.predict(x_test)
    
    #Scoring
    error=mean_squared_error(preds,y_test,squared=False)
    
    return error
#Study
study=optuna.create_study(direction='minimize')
study.optimize(objective,n_trials=10)
print("Return the best trial:",study.best_trial)
```
